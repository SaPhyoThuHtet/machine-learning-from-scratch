```python
import pandas as pd
import numpy as np

class LogisticRegression():

  def __init__(self, learning_rate, num_iterations):
    self.learning_rate = learning_rate
    self.num_iterations = num_iterations

  

  def fit(self, X, Y):
      self.X = X
      self.Y = Y
      self.W = np.zeros(self.X.shape[1])
      self.b = 0

      for i in range(self.num_iterations):
        Y_predicted = np.dot(X, self.W) + self.b
        print(Y_predicted)
        dW = np.dot(self.X.T, Y_predicted-self.Y)/self.X.shape[0]
        db = np.sum(Y_predicted-self.Y)/X.shape[0]
        self.W -= self.learning_rate*dW
        self.b -= self.learning_rate*db

  def predict(self, X):
    x = np.dot(X, self.W) + self.b
    Z= (1/(1+np.exp(-x)))
    Y = []
    for i in Z:
      if (i>=0.5):
        Y.append(1)
      else:
        Y.append(0)

    return Y
    
  
```

```python
from sklearn import datasets
iris_data = pd.read_csv('/Iris Two Class Data.csv')
iris_data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(iris_data.iloc[:,1:], iris_data.iloc[:,0], test_size = 0.25, random_state = 0)
```

```python
model = LogisticRegression(0.05,   100)
model.fit(X_train, Y_train)
```

```python
Y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score

print('Accuracy Score',accuracy_score(Y_test, Y_predicted)*100,'%')
print('Precision Macro Score',precision_score(Y_test, Y_predicted,average = 'macro')*100,'%')
print('Recall_Score',recall_score(Y_test, Y_predicted, average = 'macro')*100,'%')
```
